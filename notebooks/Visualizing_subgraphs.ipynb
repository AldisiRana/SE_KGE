```python
from seffnet.utils import create_subgraph
from seffnet.constants import DEFAULT_WEIGHTED_FULLGRAPH_PICKLE
from pybel_jupyter import to_jupyter
import pybel
```

```python
parkinson_val = create_subgraph(
    fullgraph_path=DEFAULT_WEIGHTED_FULLGRAPH_PICKLE,
    source_identifier='3121',
    source_type='chemical',
    target_name="Parkinson's disease",
    target_identifier='C0030567',
    target_type='phenotype',
    weighted=True,
    common_targets=True,
)
```

```python
to_jupyter(parkinson_val)
```
